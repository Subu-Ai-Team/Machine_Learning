```python
data.rename(columns={'Unnamed: 0'     : 'id',
                     'Battery capacity (mAh)' : 'BatteryCapacity',
                     'Screen size (inches)' : 'ScreenSize',
                     'Resolution x'   : 'Resolution_x',
                     'Resolution y'    : 'Resolution_y',
                     'RAM (MB)'  : 'RAM',
                     'Internal storage (GB)'   : 'GB',
                     'Rear camera' : 'RearCamera',
                     'Front camera'   : 'FrontCamera',
                     'Operating system'  : 'OperatingSystem',
                     'Number of SIMs'  : 'Number_of_SIMs',
                     '4G/ LTE'  : '4G/LTE'}, inplace = True)
data.head(3)
```

Pie Charts 

```python
def pie_visualization(dataframe,column):
    dataframe[column].value_counts().plot.pie(legend=True,autopct="%1.3f%%")
```

```python
pie_visualization(data,"Touchscreen")
```

```python
pie_visualization(data,"OperatingSystem")
```

```python
def plotBar(degisken, n=5):
    """
        Girdi: Değişken/sütun ismi
               n= Gösterilecek eşsiz değer sayısı
        Çıktı: Çubuk Grafiği
    """
    data_ =data[degisken]
    data_sayma = data_.value_counts()
    data_sayma = data_sayma[:n]
    plt.figure()
    plt.bar(data_sayma.index, data_sayma , color= "purple")
    plt.xticks(data_sayma.index, data_sayma.index.values)
    plt.xticks(rotation=40) #yazı eğikliği 
    plt.ylabel("Frekans")
    plt.title("Veri Sıklığı - {}".format(degisken))
    plt.show()
    print("{}: \n {}".format(degisken, data_sayma))
```

```python
kategorik_degisken = ["Brand", "Model", "Touchscreen", "OperatingSystem"]
for i in kategorik_degisken:
    plotBar(i)
```

```python
plotBar("Number_of_SIMs")
```

```python
sns.lineplot(data=data, x="BatteryCapacity", y="Price", hue="OperatingSystem", palette="BuPu")
```

```python
def plotHistogram(degisken):
    """
        Girdi: Değisken / sütun ismi
        Çıktı: İlgili değişkenin histogramı
    """
    plt.figure()
    plt.hist(data[degisken], bins=75, color = "pink")
    plt.xlabel(degisken)
    plt.ylabel("Frekans")
    plt.title("Veri Sıklığı - {}".format(degisken))
    plt.show()
```

```python
sayisal_degisken = ["BatteryCapacity", "ScreenSize", "RAM", "GB","Number_of_SIMs"]
for i in sayisal_degisken:
    plotHistogram(i)
```
